```python
import twint
import nest_asyncio
import pandas as pd
nest_asyncio.apply()
```

```python
c = twint.Config()
c.Search = "depression"
c.Format = "Username: {username} |  Tweet: {tweet}"
c.Limit = 10_000
c.Pandas = True
c.Hide_output = True
# c.Store_csv = True
# c.Output = "test_1.csv"

twint.run.Search(c)
```

```python
df = twint.output.panda.Tweets_df[["date", "username", "tweet", "hashtags", "nlikes"]]
df
```

```python
df.to_csv("tweets.csv", index=False)
```
