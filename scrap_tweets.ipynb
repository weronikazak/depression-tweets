```python
import twint
import nest_asyncio
import pandas as pd
nest_asyncio.apply()
```

## Retrieve tweets about depression

```python
c = twint.Config()
c.Search = "depression"
c.Format = "Username: {username} |  Tweet: {tweet}"
c.Limit = 10_000
c.Lang = 'en'
c.Pandas = True
c.Hide_output = True

twint.run.Search(c)
```

```python
df = twint.output.panda.Tweets_df[["date", "username", "tweet", "hashtags", "nlikes"]]
df
```

```python
df.tweet = df.tweet.str.replace("\n", " ")
df.head()
```

```python
df.to_csv("tweets_depression.csv", index=False, encoding='utf-8-sig')
```
